```python
import matplotlib.pyplot as plt
import numpy as np

# Read in the frequency data from file
data = np.loadtxt('/content/out (2).frq', skiprows=1, usecols=(4,))

# Create histogram of allele frequencies
bins = np.linspace(0, 1, num=51)
hist, _ = np.histogram(data, bins=bins)

# Plot the allele frequency spectrum
plt.bar(bins[:-1], hist, width=0.02)
plt.xlabel('Minor allele frequency')
plt.ylabel('Count')
plt.title('Allele frequency spectrum')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Read in the frequency data from file
data = np.loadtxt('/content/out (2).frq', skiprows=1, usecols=(4,))

# Create histogram of allele frequencies
bins = np.linspace(0, 1, num=51)
hist, _ = np.histogram(data, bins=bins)

# Plot the allele frequency spectrum with a logarithmic y-axis scale
plt.bar(bins[:-1], hist, width=0.02)
plt.xlabel('Minor allele frequency')
plt.ylabel('Count')
plt.title('Allele frequency spectrum')
plt.yscale('log')
plt.show()
```
